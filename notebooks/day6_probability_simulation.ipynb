```python
import numpy as np
coin = np.random.choice(["H","T"])
coin
```

```python
tosses = np.random.choice(["H","T"],size = 1000)
heads = np.sum(tosses == "H")
tails = np.sum(tosses == "T")
print("Heads:",heads)
print("Tails:",tails)
```

```python
heads/len(tosses)
```

```python
import matplotlib.pyplot as plt
counts = [heads,tails]
labels = ["Heads","Tails"]
plt.bar(labels ,counts)
plt.title("Coin toss result")
plt.show()
```
