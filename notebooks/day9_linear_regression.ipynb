```python
# Predict marks based on study hours using Linear Regression.
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,5])
y = np.array([40,50,60,70,80])
```

```python
# What linear regresion assums:
# Linear relationship
# Continuos target
# Errors minimized
```

```python
# linear regression without using libraries
m = 0
c = 0
lr = 0.01 #learning rate

for i in range(1000):
    y_pred = m*X+c
    dm = (-2/len(X))*np.sum(X*(y-y_pred))
    dc = (-2/len(X))*np.sum(y-y_pred)

    m = m-lr*dm
    c = m-lr*dc

print("Slope:",m)
print("Intercept:",c)
```

```python
predicted = m*X+c
print(predicted)
```

```python
#Visualization
plt.scatter(X,y)
plt.plot(X,predicted)
plt.xlabel("Study hours")
plt.ylabel("Marks")
plt.title("Linear regression fit")
plt.show()
```

```python
# Slope m: slope tells how output changes when input increases by unit 1.
# intercept c: intercept is predicted output wen input is zero.
# Gradient descent:method to find best m and c by reducing error step by step
```

```python
from sklearn.linear_model import LinearRegression
X2 = X.reshape(-1,1)

model = LinearRegression()
model.fit(X2,y)

print(model.coef_)
print(model.intercept_)
print(model.predict([[6]]))
```

```python
# Model Evaluation
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
y_pred = model.predict(X2)

mae = mean_absolute_error(y,y_pred)
rmse = np.sqrt(mean_squared_error(y,y_pred))

print("MAE:",mae)
print("RMSE:",rmse)
```

```python
# MAE indicates average error in predictions
# if RMSE is 2 , then model will predict marks with ±2 marks.
```
