```python
from sklearn.metrics import mean_absolute_error
y_true = [50,60,70]
y_pred = [55,58,65]
mae = mean_absolute_error(y_true,y_pred)
print(mae)
```

```python
# mae tells on average, how many marks is the model off.
# since mae is 4 onaverage,predictions are off by 4 marks.
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_true,y_pred)
print(mse)
# MSE punishes big mistakes more
# Because errors are squared, large prediction errors influence the model much more than small ones.
```

```python
import numpy as np
rmse = np.sqrt(mse)
print(rmse)

# RMSE is preferred when large errors are costly.
#RMSE is useful when large errors are more harmful than small ones because squaring the error penalizes large deviations more strongly
```
