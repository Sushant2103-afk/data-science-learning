```python
import numpy as np
import pandas as pd
data = [40, 42, 45, 47, 49, 50, 1000]
np.mean(data)
```

```python
np.median(data)
```

```python
data1 = [48, 49, 50, 51, 52]
data2 = [10, 30, 50, 70, 90]
print("mean1:",np.mean(data1))
print("standard1:",np.std(data1))
print("mean2:",np.mean(data2))
print("standard2:",np.std(data2))
```

```python
print("variance1:",np.var(data1))
print("standard1:",np.std(data1))
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

np.corrcoef(x, y)
```

```python
y2 = [10, 3, 7, 2, 9]
np.corrcoef(x, y2)
```
