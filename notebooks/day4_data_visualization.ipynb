```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
data = {
    "student": ["A", "B", "C", "D", "E"],
    "marks": [78, 45, 67, 89, 56],
    "attendance": [85, 60, 75, 95, 70]
}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['student'],df['marks'])
plt.xlabel('student')
plt.ylabel('marks')
plt.title('marks of students')
plt.show()
```

```python
# 1.highest marks are scored by student D
# 2.lower marks are scored by student B
```

```python
plt.scatter(df['attendance'],df['marks'])
plt.xlabel("attendance (%)")
plt.ylabel("marks")
plt.title("Attendance vs marks")
plt.show()
```

```python
plt.plot(df['student'],df['marks'],marker = 0)
plt.xlabel('Student')
plt.ylabel('marks')
plt.title('Mark Trend')
plt.show()
```

```python
sns.barplot(x='student',y='marks',data=df)
plt.title('Marks by student')
plt.show()
```

```python
df['result'] = df['marks'].apply(lambda x: "pass" if x>=50 else "fail")
```

```python
sns.countplot(x='result',data=df)
plt.title('pass vs fail count')
plt.show()
```
