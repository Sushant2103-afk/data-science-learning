```python
data1 = {
    "student": ["A", "B", "C", "D", "E"],
    "marks": [78, 45, None, 89, 60],
    "subject": ["Math", "Math", "Math", "Math", "Math"]
}
df1 = pd.DataFrame(data1)
df1
```

```python
df1['marks'] = df1['marks'].fillna(df1['marks'].mean(),inplace = True)
df1['marks']
```

```python
import numpy as np
df1['result'] = np.where(df1['marks'] >= 50,'pass','fail')

df1
```

```python
print("Averge_marks:",np.mean(df1['marks']))
```

```python
pass_percentage = ((df1['marks']>=50).mean()*100)
print(pass_percentage)
```
