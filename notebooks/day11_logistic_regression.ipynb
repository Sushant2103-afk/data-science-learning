```python
import pandas as pd
data = {
    "study_hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "pass":        [0, 0, 0, 1, 1, 1, 1, 1]
}
df = pd.DataFrame(data)
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
x = df[["study_hours"]]
y = df["pass"]   
model = LogisticRegression()
model.fit(x,y)
model.predict_proba([[3]])
# Output: [probability of fail, probability of pass]
```

```python
model.predict([[3]])
```

```python
from sklearn.metrics import accuracy_score
preds = model.predict(x)
accuracy = accuracy_score(y,preds)
print("Accuracy:",accuracy)
# it tells What percentage of predictions were correct.
```

```python
# The .fit() method trains the model by learning parameters from the training data, enabling it to make predictions.
```
