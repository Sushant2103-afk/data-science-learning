```python
import pandas as pd
data = {
    "student": ["A", "B", "C", "D", "E", "F", "G"],
    "marks": [78, 45, 67, 89, 56, 92, 38],
    "attendance": [85, 60, 75, 95, 70, 98, 50],
    "study_hours": [3, 1, 2, 4, 2, 5, 1]
}
df = pd.DataFrame(data)
#understanding data 
df.head()
```

```python
#checking missing values
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# univariate analysis using histplot
sns.histplot(df['marks'],bins=5,kde=True)
plt.show()
```

```python
# Bivariate analysis
# attendance vs marks
sns.scatterplot(x='attendance',y='marks',data=df)
plt.xlabel('attendace(%)')
plt.ylabel('marks')
plt.show()
```

```python
# Marks vs Study hours
sns.scatterplot(x='marks',y='study_hours',data=df)
plt.xlabel('marks')
plt.ylabel('study_hours')
plt.show()
```

```python
# Correlation
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
# Insights
# 1.Students with higher study hours tends to more score.
# 2.Attnedace also shows positive relation ,but it not only factor.
# 3.One student has low attendace and low marks.
```
